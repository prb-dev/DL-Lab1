1. Create a random 4x4 array from an exponential distribution





```python
import numpy as np

array = np.random.exponential(scale=1, size=(4, 4))
print(f'4 * 4 array: {array}')
```

2. Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions

```python
import matplotlib.pyplot as plt

plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="blue",
label="rand")
plt.axis([-0.5, 1.5, 0.9, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.Blues, linewidth=0.1)
plt.show()
```
